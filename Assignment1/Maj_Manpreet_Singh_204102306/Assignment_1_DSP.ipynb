# Generate Triangular Signal

```python
def triangularfunc(t):
    if t>= -1 and t<=0:
        return (t+1)
    elif t>=0 and t<=1:
        return (-t+1)
    else:
        return 0
tri=[]
for time in t:
    tri.append(triangularfunc(time))
    
plt.plot(t,tri)
plt.title('Triangular Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
```

# Generate Sinc Signal

```python
import math
t = np.linspace(-4,4,1000)
def sinc(t):
    return math.sin(math.pi*t)/(math.pi*t)
y =[]
for i in t:
    y.append(sinc(i))
plt.plot(t,y)
plt.title('Sinc Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
```

# 3.Systems (Time domain approach) Convolution


```python
def convolution(inputSeq, impulseResponse):
    output = []
    N1 = len(inputSeq)
    N2 = len(impulseResponse)
    N = N1+N2-1
    y = np.zeros(N)
    y1 = np.convolve(inputSeq,impulseResponse)
    m = N-N1
    n = N-N2
    #Padding zeros to inputSeq and impulseResponse to make their length to N
    inputSeq =np.pad(inputSeq,(0,m),'constant')
    impulseResponse=np.pad(impulseResponse,(0,n),'constant')
    #Linear convolution using convolution sum formula
    for n in range (N):
        for k in range (N):
            if n >= k:
                 y[n] = y[n]+inputSeq[n-k]*impulseResponse[k]
        output.append(y[n])
    return output
```

```python
x= [2,4,5,2,7]
y= [8,-5,4]
Output = convolution(x,y)
print(Output)
```
