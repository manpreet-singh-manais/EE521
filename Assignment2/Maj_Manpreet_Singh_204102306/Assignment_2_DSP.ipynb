# Uniform PDF

```python
def uniform(a,b):
    x = np.linspace(0,40,100)
    y=[]
    for i in x:
        if (a<=i<=b):
            y.append(1/(b-a))
        else:
            y.append(0)
    mean = (a+b)/2
    median = (a+b)/2
    variance = (1/12)*pow((b-a),2)
    return(x,y,mean,median,variance)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
t,uniform1,mean,median,variance= uniform(3,4)
style.use("ggplot")
plt.title("Uniform PDF with a =5 , b = 8")
plt.plot(t,uniform1,color="red")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label= 'median')
plt.plot(variance,marker="o",label = "variance")
plt.legend()

t,uniform1,mean,median,variance= uniform(2,3)
plt.subplot(2,2,2)
style.use("ggplot")
plt.title("Uniform PDF with a =4 , b = 6")
plt.plot(t,uniform1,color="purple")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label= 'median')
plt.plot(variance,marker="o",label = "variance")
plt.legend()

t,uniform1,mean,median,variance= uniform(1,2)
plt.subplot(2,2,3)
style.use("ggplot")
plt.plot(t,uniform1,color="blue")
plt.title("Uniform PDF with a =3 , b = 8")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label= 'median')
plt.plot(variance,marker="o",label = "variance")
plt.legend()
plt.show()
```

# Exponential PDF

```python
def exponential(L):
    x = np.linspace(-10,40,100)
    y=[]
    for i in x:
        if (i>0):
            y.append(L * np.exp(-L*i))
        else:
            y.append(0)
    mean = 1/L
    median = np.log(2)/L
    variance = 1/pow(L,2)
    return(x,y,mean,median,variance)
```

```python
plt.figure(figsize=(15,8))
t,exponential1,mean,median,variance =  exponential(0.89)
plt.subplot(2,2,1)
style.use("ggplot")
plt.plot(t,exponential1,color="pink")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label= 'median')
plt.plot(variance,marker="o",label = "variance")
plt.title("Exponential PDF with lambda = 0.15")
plt.legend()

t,exponential1,mean,median,variance =  exponential(0.95)
style.use("ggplot")
plt.subplot(2,2,2)
plt.plot(t,exponential1,color="orange")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label= 'median')
plt.plot(variance,marker="o",label = "variance")
plt.title("Exponential PDF with lambda = 0.95")
plt.legend()

t,exponential1,mean,median,variance =  exponential(1)
style.use("ggplot")
plt.subplot(2,2,3)
plt.plot(t,exponential1,color="brown")
plt.title("Exponential PDF")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label= 'median')
plt.plot(variance,marker="o",label = "variance")
plt.title("Exponential PDF with lambda = 1")
plt.legend()
plt.show()
```

# Gamma PDF

```python
def Gamma(a,L):
    x = np.linspace(-10,40,100)
    y=[]
    for i in x:
        if (i>0):
            y.append((pow(L,a)*pow(i,(a-1))*np.exp(-L*i))/math.gamma(a))
        else:
            y.append(0)
    mean = (a/L)
    variance = (a/pow(L,2)) 
    return(x,y,mean,variance)    
        
```

```python
plt.figure(figsize=(15,8))
t,gamma1,mean,variance =  Gamma(2,1.5)
plt.subplot(2,2,1)
style.use("ggplot")
plt.plot(t,gamma1,color="pink")
plt.plot(mean,marker="^",label = "mean")
plt.plot(variance,marker="o",label = "variance")
plt.title("Gamma PDF with lambda = 1.5, Alpha = 2")
plt.legend()

t,gamma1,mean,variance =  Gamma(4.5,3)
plt.subplot(2,2,2)
style.use("ggplot")
plt.plot(t,gamma1,color="orange")
plt.plot(mean,marker="^",label = "mean")
plt.plot(variance,marker="o",label = "variance")
plt.title("Gamma PDF with lambda = 3, Alpha = 4.5")
plt.legend()

t,gamma1,mean,variance =  Gamma(3.25,5)
plt.subplot(2,2,3)
style.use("ggplot")
plt.plot(t,gamma1,color="violet")
plt.plot(mean,marker="^",label = "mean")
plt.plot(variance,marker="o",label = "variance")
plt.title("Gamma PDF with lambda = 5, Alpha = 13.25")
plt.legend()
plt.show()
```

# Beta PDF

```python
def Beta(a,b):
    x = np.linspace(-1,1.5,100)
    y=[]
    for i in x:
        if (0<=i<1):
            y.append(((math.gamma(a+b)*pow(i,(a-1))*pow((1-i),(b-1))))/(math.gamma(a)*math.gamma(b)))
        else:
            y.append(0)
            
    mean = (a/(a+b))
    median = (a-1/3)/(a+b-2/3)
    mode = (a-1)/(a+b-2)
    variance = (a*b/(pow((a+b),2)*(a+b+1)))
    return(x,y,mean,median,mode,variance)  
```

```python
plt.figure(figsize=(15,8))
t,beta1,mean,median,mode,variance= Beta(4,15)
plt.subplot(2,2,1)
style.use("ggplot")
plt.plot(t,gamma1,color="red")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label = "median")
plt.plot(mode,marker=">",label = "Mode")
plt.plot(variance,marker="o",label = "variance")
plt.title("Beta PDF with a = 2, b= 23")
plt.legend()

t,beta1,mean,median,mode,variance= Beta(6,15)
plt.subplot(2,2,2)
style.use("ggplot")
plt.plot(t,gamma1,color="green")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label = "median")
plt.plot(mode,marker=">",label = "Mode")
plt.plot(variance,marker="o",label = "variance")
plt.title("Beta PDF with a = 4, b=23")
plt.legend()


t,beta1,mean,median,mode,variance= Beta(2,15)
plt.subplot(2,2,3)
style.use("ggplot")
plt.plot(t,gamma1,color="purple")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label = "median")
plt.plot(mode,marker=">",label = "Mode")
plt.plot(variance,marker="o",label = "variance")
plt.title("Beta PDF with a = 5, b= 23")
plt.legend()
plt.show()
```

# Gaussian PDF

```python
def Gaussian(sigma,mu):
    x = np.linspace(-30,30,100)
    y=[]
    for i in x:
            y.append((1/(sigma*np.sqrt(2*np.pi)))*(np.exp(((-1)*pow((i-mu),2))/(2*pow(sigma,2)))))
    mean = mu
    median = mu
    mode = mu
    variance = pow(sigma,2)
    return(x,y,mean,median,mode,variance)  

```

# Chi-Squared PDF

```python
def chi_square(k):
    x = np.linspace(-2,30,100)
    y=[]
    for i in x:
        if (i >= 0):
            y.append((pow(i,((k/2)-1))*np.exp(-i/2))/((pow(2,k/2))*math.gamma(k/2)))
        else:
            y.append(0)
    mean = k
    median = k*pow((1-2/(9*k)),2)
    variance = 2*k
    return(x,y,mean,median,variance)  
```

```python
plt.figure(figsize=(15,8))
t,chi_square1,mean,median,variance= chi_square(0.5)
plt.subplot(2,2,1)
style.use("ggplot")
plt.plot(t,chi_square1,color="red")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label = "median")
plt.plot(variance,marker="o",label = "variance")
plt.title("Gaussian PDF with k= 0.5")
plt.legend()

t,chi_square1,mean,median,variance= chi_square(1)
plt.subplot(2,2,2)
style.use("ggplot")
plt.plot(t,chi_square1,color="red")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label = "median")
plt.plot(variance,marker="o",label = "variance")
plt.title("Gaussian PDF with k= 1")
plt.legend()


t,chi_square1,mean,median,variance= chi_square(1.5)
plt.subplot(2,2,3)
style.use("ggplot")
plt.plot(t,chi_square1,color="red")
plt.plot(mean,marker="^",label = "mean")
plt.plot(median,marker="*",label = "median")
plt.plot(variance,marker="o",label = "variance")
plt.title("Gaussian PDF with k= 1.5")
plt.legend()
plt.show()
```

# Discrete distributions

```python
"""
Consider a transmitter that can transmit messages only in the form
of 2 bits 0 and 1. The receiver receives a single message sent by the
transmitter say X. Since, we do not know before hand which bit
was sent X follows a certain distribution. Which distribution is this?
Find the PMF and plot it. Also calculate the mean and variance of
X. Assume probability for receiving 0(failure) is p = 0.1, 0.5, 0.8
"""
def bernouli(p):
    b=[]
    for i in x:
        if (i==0):
            b.append((1-p))
        if i==1:
            b.append(p)
    return b
x=[0,1]
plt.figure(figsize=(9,12))
plt.subplot(3,1,1)
style.use("ggplot")
t1=bernouli(0.1)
plt.stem(x,t1,use_line_collection=True)
plt.title('Bernoulli distribution with values of p = 0.1  and q= 0.9')
plt.subplot(3,1,2)
style.use("ggplot")
t2=bernouli(0.5)
plt.stem(x,t2,use_line_collection=True)
plt.title('Bernoulli distribution with values of p = 0.5 and q= 0.5')
plt.subplot(3,1,3)
style.use("ggplot")
t3=bernouli(0.8)
plt.title('Bernoulli distribution with Values of p = 0.8 and q= 0.2')
plt.stem(x,t3,use_line_collection=True)
plt.show()
```

```python
"""
Consider the same transmitter but this time the receiver receives 20
bits. Each bit is independent of the previous one. The probability of
receiving a 0 is p = 0.4. Let X be the number of 1s (successes) that
is received at the receiver’s end. What is the distribution followed
by X. How is it related to the first distribution? Calculate the mean
and variance and plot the PMF.
"""
n=20
r=np.arange(0,n+1)
def binomial(n,p):
    y1=[]
    for r in range(n+1):
        y=(math.factorial(n)/(math.factorial(n-r)*math.factorial(r)))*(p**r)*((1-p)**(n-r))
        y1.append(y) 
    return (y1) 
t1=binomial(20,0.4)
plt.stem(r,t1,use_line_collection=True)
plt.show()
```

# Autocorrelation

```python
def convolution(inputSeq, impulseResponse):
    output = []
    N1 = len(inputSeq)
    N2 = len(impulseResponse)
    N = N1+N2-1
    y = np.zeros(N)
    y1 = np.convolve(inputSeq,impulseResponse)
    m = N-N1
    n = N-N2
    #Padding zeros to inputSeq and impulseResponse to make their length to N
    inputSeq =np.pad(inputSeq,(0,m),'constant')
    impulseResponse=np.pad(impulseResponse,(0,n),'constant')
    #Linear convolution using convolution sum formula
    for n in range (N):
        for k in range (N):
            if n >= k:
                 y[n] = y[n]+inputSeq[n-k]*impulseResponse[k]
        output.append(y[n])
    return output

def autocorrelation(seq1,delay):
    for i in range(0,delay):
        seq1[:0] = [0]
    a = seq1
    b = seq1[::-1]
    autocorrelation = convolution(a,b)
    return autocorrelation
```

```python
#Plotting the Autocorrelation:

plt.figure(figsize=(12,20))
t1 = autocorrelation(x1,2)
t2 = autocorrelation(x2,2)
t3 = autocorrelation(x3,2)
t4 = autocorrelation(x4,2)
t5 = autocorrelation(x5,2)
t6 = autocorrelation(x6,2)

plt.subplot(6,1,1)
style.use("ggplot")
plt.title('x11')
plt.stem(t1,use_line_collection=True)
plt.subplot(6,1,2)
style.use("ggplot")
plt.title('x22')
plt.stem(t2,use_line_collection=True)
plt.subplot(6,1,3)
style.use("ggplot")
plt.title('x33')
plt.stem(t3,use_line_collection=True)
plt.subplot(6,1,4)
style.use("ggplot")
plt.title('x44')
plt.stem(t4,use_line_collection=True)
plt.subplot(6,1,5)
style.use("ggplot")
plt.title('x55')
plt.stem(t5,use_line_collection=True)
plt.subplot(6,1,6)
style.use("ggplot")
plt.title('x66')
plt.stem(t6,use_line_collection=True)
style.use("ggplot")
```
